## Modelado de mecánica de una esfera a traves de la tierra

### El fenómeno que se pretende modelar, consiste en un sistema en el que se supone un planeta de proporciones similares a las de la tierra, donde este planeta posee la misma gravedad y un tunel cilíndrico que lo recorre de polo a polo, la pregunta que se hizo el que formuló este problema fué, cuanto tarda una esfera en atravesar el planeta y volver, considerando las variaciones de la dirección de la gravedad, por lo tanto se establecieron un conjunto de condiciones para este modelo:

- La gravedad es igual a la de la tierra.
- Se desprecia la fricción en las paredes internas del tunel.
- Se desconoce la masa del objeto a lanzar.
- Se estipula que el tunel cruza de lado a lado sin obstáculo alguno en el interior.
- Se asume que la forma de la tierra es perfectamente esférica.
- Se asume que en el centro de la tierra hay temperaturas soportables por la esfera.

## Definición de variables
- g: aceleración de la gravedad variable en el tiempo
- ge: aceleración de la gravedad como parámetro $= 9.8 m/s^2$
- M: masa de la tierra
- m: masa de la tierra variable
- p: densidad de la tierra.
- r: radio de la tierra variable.
- G: constante gavitacional universal $= 6.674 * e^{-11}$
- Re: radio real de la tierra.
- w: velocidad angular.
- T: periodo.
- k: constante de elasticidad.
- T: tiempo

```python
#El usuario debe ingresar la masa de la esfera como parámetro de entrada 

def simulated(masa):
    import numpy as np
    import matplotlib.pyplot as plt
    g=9.8           #Gravedad
    Me=5972e24     #Masa de la Tierra
    ConsG=6674e-11 #Constante Gravitacional
    RadioT=6371     #Radio de la Tierra1
    posLog=[]       #Log de posición
    velLog=[]       #Log de Velocidad
    aceLog=[]       #Log de Aceleración
    t=0
    k=(masa*g)/(RadioT*1000)
    print("Inicio de Simulación T=0")
    #x=-RadioT
    for i in range(1,2550):
        x=RadioT*np.cos(np.sqrt(k/masa)*t)
        vel=-RadioT*(np.sqrt(k/masa))*(np.sin((np.sqrt(k/masa)*t)))
        ace=((-RadioT*k)/masa)*(np.cos(np.sqrt(k/masa)*t))
        posLog.append(x)
        velLog.append(vel)
        aceLog.append(ace)
        print(f'i= {i} tnow = {t}  Pos = {x} Vel = {vel} Ace = {ace} ')
        t+=1
    #Imprimimos el traceo de datos de las colas
    plt.figure(1,figsize=(15,5))
    plt.subplot(221)
    plt.plot(posLog ,label='Posición')
    plt.title('Posición')
    plt.xlabel('Tiempo')
    plt.ylabel('Posición')
    plt.grid()
    
    plt.subplot(222)
    plt.plot(velLog ,label='Velocidad')
    plt.legend()
    plt.title('Velocidad')
    plt.xlabel('Tiempo')
    plt.ylabel('Velocidad')
    plt.grid()
    
    plt.subplot(223)
    plt.plot(aceLog,label='Aceleración')
    plt.legend()
    plt.title('Acleración')
    plt.xlabel('Tiempo')
    plt.ylabel('Aceleración')
    plt.grid()
   
    
    
    plt.show()
    
simulated(1000)
```
